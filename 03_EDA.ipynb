## Add Billboard Top 100 .csv file to DF and create field BB_Hot to determine if a song was popular or not for later modeling

```python
billboard = pd.read_csv("./Data/billboard.csv")
billboard.head()
```

```python
#Creating a function that takes a list of the track names from the BillBoard top 100 that was
#webscraped and look for a matching title in the MSD subset.  If it finds a matching title, this
#means it was a popular song, and creating a binary column 1=True, 0=False to track this for modeling purposes

billboard_tracks = list(billboard.track)

def match_title(row):
    bb_list = billboard_tracks
    if row['title'] in bb_list:
        return 1
    else:
        return 0
```

```python
#add to df
df['bb_hot'] = df.apply(match_title, axis=1)
```
